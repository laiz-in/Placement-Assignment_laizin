```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
#import neccessaryy libraries 
```

```python
#read the json data from given link
data_url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
response = requests.get(data_url)
data = response.json()

# Convert the data into a DataFrame
df = pd.DataFrame(data['pokemon'])
```

# Get all Pokemons whose spawn rate is less than 5%

```python
low_spawn_rate_pokemon = df[df['spawn_chance'] < 5]
low_spawn_rate_pokemon
```

```python
plt.figure(figsize=(20, 6))
plt.bar(low_spawn_rate_pokemon['name'], low_spawn_rate_pokemon['spawn_chance'])
plt.title('Pokémon with Spawn Chance < 5%')
plt.xlabel('Pokémon')
plt.ylabel('Spawn Chance')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# ● Get all Pokemons that do not have more than 2 evolutions


```python
df
```

```python
# Calculate the total number of evolutions for each Pokémon
df['total_evolutions'] = df.apply(lambda x: len(x['prev_evolution']) if isinstance(x['prev_evolution'], list) else 0, axis=1) + \
                          df.apply(lambda x: len(x['next_evolution']) if isinstance(x['next_evolution'], list) else 0, axis=1)

# Filter the Pokémon that have less than or equal to 2 evolutions
LessThan_3_evolutions = df[df['total_evolutions'] <= 2]


pd.set_option('display.max_colwidth', None)
LessThan_3_evolutions
```

* we are adding up the next evoultion and previous evolution to find the total mumner of evolution of a pokemon
* if it doesnt have a prev evolution or next evolution , we are considering it as zero

# below plot does not include pokemons with more than 2 evolutions

```python
plt.figure(figsize=(20, 6))
plt.bar(LessThan_3_evolutions['name'], LessThan_3_evolutions['total_evolutions'])
plt.title('Pokémon with Less than 3 Evolutions')
plt.xlabel('Pokémon')
plt.ylabel('Number of Evolutions')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#  Get all Pokemons whose spawn time is less than 300 seconds

```python
spawntime= df[['name','spawn_time']].copy()
spawntime.columns = ['name', 'spawn_time']

spawntime
```

```python
spawntime['spawn_time'].replace("N/A","00:00", inplace=True)
#Replacing the N/A columnn with 00:0 format so that it will consider it as zero seconds
pd.reset_option('display.max_rows')

spawntime
```

```python
time_components = spawntime['spawn_time'].str.split(':')

# Convert minutes and seconds to integers
minutes = time_components.str[0].astype(int)
seconds = time_components.str[1].astype(int)

# Calculate the spawn time in seconds and add it as a new column
spawntime['seconds'] = minutes * 60 + seconds
spawntime
```

```python
less_than_300=spawntime[spawntime['seconds']<300]
greater_than_300=spawntime[spawntime['seconds']>=300]
```

```python
plt.figure(figsize=(20, 6))
plt.bar(greater_than_300['name'], greater_than_300['seconds'],color='skyblue')
plt.bar(less_than_300['name'], less_than_300['seconds'], color='salmon')  # Different color for selected bars

plt.xlabel('Pokémon')
plt.ylabel('Spawn Time (seconds)')
plt.title('Pokémon Spawn Time (Less than 300 seconds)')
plt.xticks(rotation=90)
plt.tight_layout()

# Display the plot
plt.show()
#skyblue shows the names of pokemon which are having spawn time greater than or equals to 300 seconds
#salmon represents the names of pokemon which are having the spawn time less than 300
```
