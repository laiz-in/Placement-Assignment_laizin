# Get all the Earth meteorites that fell before the year 2000

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
#import neccessaryy libraries 
```

```python
data_url = "https://data.nasa.gov/resource/y77d-th95.json"
response = requests.get(data_url)
data = response.json()
df = pd.DataFrame(data)
data
```

```python
df=df.drop(':@computed_region_cbhk_fwbd',axis=True)
df
```

```python
df=df.drop(':@computed_region_nnqa_25f4',axis=True)
df
```

```python
df
```

● Get all the Earth meteorites that fell before the year 2000


```python
#to get this done , w are converting the year column to frist 4 letters only , that represents the year
fellYear= df[['name','year']].copy()
fellYear.columns = ['name', 'year']
```

```python
fellYear['year']=fellYear['year'].str[:4]
#taking out the fist 4 letters that represents the year

#droppin gthe rows with nan values
fellYear = fellYear.dropna(subset=['year'])

#converting into integers
fellYear['year'] = fellYear['year'].astype(int)



fellYear
```

```python
before_2000 = fellYear[fellYear['year'] < 2000]

# Get the list of names
names_before_2000 = before_2000['name'].tolist()
names_before_2000
#this list contains all the meteorites fell before 2000
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(fellYear['name'], fellYear['year'], alpha=0.5)
plt.title('Meteorites vs Year')
plt.ylabel('Year that Meteorites Have Fallen')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Get all the earth meteorites co-ordinates who fell before the year 1970


```python
cordinates= df[['name','year','geolocation']].copy()
cordinates.columns = ['name', 'year','geolocation']
cordinates
```

```python
cordinates['year']=cordinates['year'].str[:4]
#taking out the fist 4 letters that represents the year

#droppin gthe rows with nan values
cordinates = cordinates.dropna(subset=['year'])

#converting into integers
cordinates['year'] = cordinates['year'].astype(int)
```

```python
cordinates
```

```python
before_1970= cordinates[cordinates['year'] < 2000]
pd.set_option('display.max_colwidth', None)
before_1970
```

```python
before_1970
```

```python
#function to extract the cordinates from the gelocation column
def extract_coordinates(row):
    if isinstance(row['geolocation'], dict):
        return row['geolocation']['coordinates']
    else:
        return None
```

```python
#creating a new column named as coordinates
before_1970['coordinates'] = before_1970.apply(extract_coordinates, axis=1)
```

```python
before_1970 = before_1970.drop('geolocation', axis=1)
#dropping the exisitng column after extracting the cordinates
```

```python
before_1970['name']
```

# Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg


```python
df
```

```python
masses= df[['name','mass']].copy()
masses.columns = ['name', 'mass']
masses
```

```python
masses = masses.dropna(subset=['mass'])

#converting into integers
masses['mass'] = masses['mass'].astype(float)
greaterThan10000= masses[masses['mass'] >10000]
greaterThan10000
```

here are the names of mateorites which weighs more than 10000 kg

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.scatter(greaterThan10000['name'], greaterThan10000['mass'])
plt.title('Meteorites vs mass')
plt.ylabel('mass of the Meteorite')
plt.xticks(rotation=90)
plt.tight_layout()

# Set y-axis ticks
y_ticks = plt.MaxNLocator(5)
plt.gca().yaxis.set_major_locator(y_ticks)

plt.show()
```
